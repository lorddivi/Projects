```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
tracks = pd.read_csv('tracks.csv')
tracks.head()
```

```python
# checking for null values
```

```python
pd.isnull(tracks).sum()
```

```python
tracks.info()
```

```python
#finding the 10 least popular songs
```

```python
sorted_tracks = tracks.sort_values(by = 'popularity',ascending = True).head(10)
sorted_tracks
```

```python
tracks.describe().T
```

```python
#top 10 songs with popularity greater than 90
```

```python
most_popular = tracks.query('popularity>90', inplace = False).sort_values('popularity',ascending = False)
most_popular[:10]
```

```python
tracks.set_index('release_date', inplace = True)
tracks.index = pd.to_datetime(tracks.index)
tracks.head()
```

```python
tracks[['artists']].iloc[18]
```

```python
# converting duration to seconds
```

```python
tracks['duration'] = tracks['duration_ms'].apply(lambda x: round(x/1000))
tracks.drop('duration_ms', inplace = True, axis = 1)
```

```python
tracks.duration.head()
```

```python
corr = tracks.drop(['key','mode','explicit'],axis = 1).corr(method = 'pearson')
plt.figure(figsize = (14,6))
heatmap = sn.heatmap(corr,annot = True,fmt = ".1g",vmin = -1,vmax =1,center = 0,cmap = 'inferno', linewidths = 1, linecolor = 'Black')
heatmap.set_title('Correlation heatmap between variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation = 90)
```

```python
sample_df = tracks.sample(int(0.004*len(tracks)))
```

```python
print(len(sample_df))
```

```python
#creating a regression plot b/w loudness and energy
```

```python
plt.figure(figsize = (10,6))
sn.regplot(data = sample_df, y = 'loudness', x = 'energy', color = 'c').set(title('Loudness vs Energy Correlation'))
```

```python
plt.figure(figsize = (10,6))
sn.regplot(data = sample_df, y = 'popularity', x = 'acousticness', color = 'b').set(title('popularity vs acousticness Correlation'))
```

```python
 tracks['dates'] = tracks.index.get_level_values('release_date')
```

```python
tracks.dates = pd.to_datetime(tracks.dates)
years = tracks.dates.dt.year
```

```python
# creating a distribution plot to visualize total no of songs since 1922
```

```python
sn.displot(years,discrete = True, aspect = 2, height = 5, kind = 'hist').set(title = 'Number of songs per year')
```

```python
total_duration = tracks.duration
fig_dim = (18,7)
fig, ax = plt.subplots(figsize = fig_dim)
fig = sn.barplot(x = years, y = total_duration, ax = ax, errwidth = False).set(title = 'Year vs Duration')
plt.xticks(rotation = 90)
```

```python
#above is a barplot showing trend in song duration over the years
```

```python
#below is a line plot showing avg duration over the years
```

```python
sn.set_style(style = 'whitegrid')
fig_dim = (10,5)
fig, ax = plt.subplots(figsize = fig_dim)
fig = sn.lineplot(x = years, y = total_duration, ax = ax,).set(title = 'Year vs Duration')
plt.xticks(rotation = 60)
```

```python
genres = pd.read_csv('SpotifyFeatures.csv')
```

```python
genres.head()
```

```python
plt.title('Duration of songs in different genres')
sn.color_palette('rocket',as_cmap = True)
sn.barplot(y = 'genre', x= 'duration_ms', data = genres)
plt.xlabel('Duration in ms')
plt.ylabel('Genres')
```

```python
sn.set_style(style = 'darkgrid')
plt.figure(figsize = (10,5))
famous = genres.sort_values(by = 'popularity',ascending = False).head(10)
sn.barplot(y = 'genre', x= 'popularity', data = famous).set(title = 'Top 5 genres by popularity')
```
